# Polynomial Regression Assignment

```python
import numpy as np
import matplotlib
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model, preprocessing
import numpy.polynomial.polynomial as poly
%matplotlib inline
```

# Importing Data 
Import data from poly_regression.csv

```python
# Run the code cell
df = pd.read_csv('poly_regression.csv')
data = df.values
X = data[:,0]
Y = data[:,1]
X = X.reshape(X.shape[0],1)
Y = Y.reshape(Y.shape[0],1)
```

# Training and Test Set
Divide the dataset approximately into 2 equal parts : training and test set. 

```python
X_tr = X[0:100,0]
Y_tr = Y[0:100,0]
X_test = X[100:200,0]
Y_test = Y[100:200,0]
```

# Plotting Dataset

```python
# Write code to plot the dataset.
plt.scatter(X_tr, Y_tr, c = 'b', s = 1)
plt.scatter(X_test,Y_test, c = 'r', s= 1)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Training Data & test Data")
```

# Model Selection.
Write code to obtain polynomial regression models of different orders starting from linear regression i.e. degree = 1 to higher degree models like degree = 2 to 10. Find training and test error for every order and plot these errors v/s degree. Select the order that fits the data best based on low training and test error. You can use poly.polyval method from numpy to find coefficients of the different models.

```python
training_err = []
test_err = []

for degree in range(1,11):
    print("With polynomial of degree: ",str(degree))
    
    w = poly.polyfit(X_tr, Y_tr, degree)
    y1 = poly.polyval(X_test,w) 
    test_err.append(np.sum((Y_test - y1)**2)/ (2 * Y_test.shape[0]))
    
    y2 = poly.polyval(X_tr,w) 
    training_err.append(np.sum((Y_tr - y2)**2)/ (2 * Y_tr.shape[0]))
    
    plt.scatter(X_tr,Y_tr,c ='b',s=1)
    plt.scatter(X_tr,y2, c='r',s=1)
    plt.xlabel('X_tr')
    plt.xlim([-1.25,1.25])

    plt.show()
    

degree = np.array(range(1,11))
plt.plot(degree,test_err,'x',c='g')
plt.plot(degree,training_err,'x',c='y')
plt.xlabel('Degree')
plt.title('Degree VS. Test/Training Error')
```
